## 그래프 시각화

#### matplotlib, numpy load

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.randn(50).cumsum()
data
```

### 점선 그래프 그리기

```python
plt.plot(data)
plt.show()
```

### 그래프 한번에 여러개 그려보기

#### 여러개 그릴 준비하기 subplot

```python
plt.subplot(1, 2, 1)
plt.subplot(1, 2, 2)
plt.show()
```

1 (행의 개수), 2 (열의 개수), 1 (나누어진 것 들 중에서 첫번째)

```python
plt.subplot(3, 2, 1)
plt.subplot(3, 2, 2)
plt.subplot(3, 2, 3)
plt.subplot(3, 2, 4)
plt.subplot(3, 2, 5)
plt.subplot(3, 2, 6)
plt.show()
```

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2, 2, 1)
plt.plot(data)
plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=20)
plt.subplot(2, 2, 3)
# plt.scatter ( x, y )
plt.scatter(scat_data, np.arange(30) + 3 * np.random.randn(30))
plt.show()
```

plt.subplot 다음에 plt.plot 이 순서가 중요함

### 그래프 선 옵션

#### plt.plot(data, '색상')

```python
plt.plot(data, 'b')
plt.show()
```

```python
plt.plot(data, 'g')
plt.show()
```

#### 색상 뒤에 마커를 같이 붙여서 옵션을 줄 수도 있음
- 색상 : b (파랑) / g (초록) / r (빨강) / c (청록) / y (노랑) / k (검정) / w (흰색)
- 마커 : o (원) / v (역삼각형) / ^ (삼각형) / s (네모) / + (플러스) / . (점)

```python
plt.plot(data, 'yv')
plt.show()
```

```python
plt.plot(data, 'b--')
plt.show()
```

```python
plt.plot(data, 'g.')
plt.show()
```

### 그래프 사이즈 조절

plt.figure 안에 figsize를 이용하여 가로, 세로 길이 조정 가능 (inch 단위)

```python
plt.figure(figsize=(10, 5))
plt.plot(data, 'rs')
plt.show()
```

#### 여러개의 그래프를 한번에 크기를 변경하려면 가장 첫 줄에 작성

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.plot(data)
plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=20)
plt.subplot(2, 2, 3)
# plt.scatter ( x, y )
plt.scatter(scat_data, np.arange(30) + 3 * np.random.randn(30))
plt.show() 
```

### 그래프 겹쳐서 그리기

subplot을 사용하지 않으면 하나의 그래프로 합쳐서 결과가 나옴

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'r--')
plt.plot(data, 'g-', drawstyle='steps-post')
plt.show()
```

### lengend 붙여주기

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'r--')
plt.plot(data, 'b-', drawstyle='steps-post', label='steps-post')
plt.legend()
plt.show()
```

### 이름 달기

```python
plt.plot(np.random.randn(1000).cumsum())
plt.title('Random Graph')
plt.xlabel('Strages')
plt.ylabel('Values')
plt.show()
```

### 종합

```python
plt.plot(np.random.randn(1000).cumsum(), 'k', label='one')
plt.plot(np.random.randn(1000).cumsum(), 'b', label='two')
plt.plot(np.random.randn(1000).cumsum(), 'r', label='three')
plt.legend()
plt.show()
```

```python
plt.title("Fianl Result")
plt.plot(np.random.randn(1000).cumsum(), 'k--', label='one')
plt.plot(np.random.randn(1000).cumsum(), 'b', label='two')
plt.plot(np.random.randn(1000).cumsum(), 'r', label='three')
plt.legend()
plt.show()
```
