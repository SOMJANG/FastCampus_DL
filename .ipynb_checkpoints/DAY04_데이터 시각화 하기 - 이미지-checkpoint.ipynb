## 이미지 시각화

### numpy, PIL, matplotlib import

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

### 이미지 파일 열기

```python
path = './images/src/'
name = ['avengers.jpg', 'marvel.jpeg']

image_pil = Image.open(path+name[0])
image = np.array(image_pil)
```

### 이미지 shape 확인
#### 734 ( 세로의 길이 ), 522 ( 가로의 길이 ), 3 ( r, g, b ) -> 색상이 있는 사진이라 마지막이 3
그레이스케일일 경우 0또는 1

```python
image.shape
```

### 이미지 range 확인
#### 혹시라도 255 이상의 값을 가지는 다른 데이터 타입을 가지는 이미지 일 수 도 있어 종종 확인

```python
np.min(image), np.max(image)
```

---

### 그래프로 시각화 하기

#### 히스토그램 그려보기
ravel로 쭉 펼친다음에 그래프로 그려보기

```python
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

### 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

### 이미지를 흑백으로 열기

```python
image_pil = Image.open(path+name[0]).convert("L")
image_bw = np.array(image_pil)
```

#### 기존에 3 ( r, g, b ) 였던 것이 -> 없어짐

```python
image.shape, image_bw.shape
```

### 흑백 이미지 열기

```python
plt.imshow(image_bw)
plt.show()
```

```python
plt.imshow(image_bw, 'gray')
plt.show()
```

### 다른 색상으로 cmap 표현하기

### RdBu : 빨강 파랑으로 표현

```python
plt.imshow(image_bw, 'RdBu')
plt.show()
```

### jet : 높을 수록 붉은 색 낮을 수록 푸른색

```python
plt.imshow(image_bw, 'jet')
plt.show()
```

### Colorbar 추가하기

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

### 이미지 설정
#### 사이즈 조절

```python
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```

#### 이미지에 제목 추가

```python
plt.title("Avengers Endgame")
plt.imshow(image)
plt.show()
```

### 새로운 이미지 열기

```python
path2 = "./images/src/"

image_pil2 = Image.open(path2+name[1])
image_2 = np.array(image_pil2)
```

```python
plt.imshow(image_2)
plt.show()
```

### 새로운 이미지를 첫번째 이미지 모양에 맞추기

#### 첫번째 이미지 : (772, 522, 3)    두번째 이미지 : (768, 1366, 3)

```python
image.shape, image_2.shape
```

### 새로운 이미지를 합치기 위한 준비

```python
import cv2
```

```python
avg_image = cv2.resize(image, (1366, 768))
avg_image.shape
```

```python
avg_image.shape, image_2.shape
```

###  이미지 합치기
그래프처럼 그냥 겹쳐서 그리면 두번째 이미지에 원래 이미지가 가려져서 안보임

```python
plt.imshow(avg_image)
plt.imshow(image_2)
plt.show()
```

#### 투명값에 관여하는 alpha 활용

```python
plt.imshow(avg_image)
plt.imshow(image_2, alpha=0.5)
plt.show()
```

### Subplot

```python
plt.figure(figsize=(30, 20))
plt.subplot(221)
plt.imshow(avg_image)
plt.subplot(222)
plt.imshow(image_2)
plt.subplot(223)
plt.imshow(image)
```
